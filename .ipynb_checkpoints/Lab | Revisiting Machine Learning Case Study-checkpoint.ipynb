```python
# Now again finding the percentage of null Values
nulls_percent_cat = pd.DataFrame(categorical.isna().sum()/len(data)).reset_index()
nulls_percent_cat
nulls_percent_cat.columns = ['column_name', 'nulls_percentage']
nulls_percent_cat
nulls_percent_cat[nulls_percent_cat['nulls_percentage']!=0]
```

## Create a new empty list called drop_list

```python
drop_list=[]
categorical['OSOURCE'].value_counts()
```

## Identify columns that over 85% missing values

```python
columns_above_threshold = nulls_percent_cat[nulls_percent_cat['nulls_percentage']>0.85]
columns_above_threshold
```

## Remove those columns from the dataframe

```python
drop_list = list(columns_above_threshold['column_name'])
drop_list
```

```python
categorical = categorical.drop(columns=drop_list)
```

```python
categorical
```

```python
#Dropping the column 'Osource' and 'Zip Code'
categorical = categorical.drop(columns=['OSOURCE','ZIP'])
```

## Reduce the number of categories in the column GENDER

```python
categorical['GENDER'].value_counts()
```

```python
categorical['GENDER'].isna().sum()
```

```python
#Filling the NaN with F as specified
categorical['GENDER'] = categorical['GENDER'].fillna('F')
categorical['GENDER'].value_counts()
```

```python
#Creating a function
def clean_gender(x):
    if x in "M":
        return "M"
    if x in "F":
        return "F"
    else:
        return "other"
```

```python
categorical['GENDER'] = list(map(clean_gender, categorical['GENDER']))
```

```python
categorical['GENDER'].value_counts()
```

```python
categorical
```
